```python
import os
import shutil
# Source and destination directories
src_dir = r"/content/datasets/AntiCheating-1/train/images"
dst_dir = r"/content/datasets/AntiCheating-1/valid/images"
# Get a list of all files in the source directory
files = os.listdir(src_dir)
# Sort the files in alphabetical order
files.sort()
# Move the first 46 files to the destination directory
for file in files[:46]:
  src_file = os.path.join(src_dir, file)
  dst_file = os.path.join(dst_dir, file)
  shutil.move(src_file, dst_file)
```

```python
import zipfile
import os
# Set the directory path for the folder you want to download
dir_path = "/content/runs"
# Set the path for the zip file you want to create
zip_path = "/content/AntiCheating_run_2.zip"
# Create a zipfile object
zipf = zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED)
# Walk through all the directories and subdirectories in the given path
for root, dirs, files in os.walk(dir_path):
  for file in files:
    # Get the absolute path of the file
    file_path = os.path.join(root, file)
    # Add the file to the zipfile object
    zipf.write(file_path)
    # Close the zipfile object
zipf.close()
# Download the zipfile
from google.colab import files
files.download(zip_path)
```
